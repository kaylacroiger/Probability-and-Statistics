#Scatterplot Array

```python
import numpy as np
import matplotlib.pyplot as plt
#importing libraries
```

```python
x = np.linspace(-10,10,20)
#creates an array with even linespaces with 20 points from the range of -10 through 10
```

```python
x
#produces the x values for the array
```

```python
m = 50
b = 17
#defines slope to equal 50 and y intercept to equal 14
```

```python
y = m * x + b
#formula to plot a linear graph using the given x, m and b values
```

```python
plt.plot(x,y,'.')
```

```python
r = 100*np.random.randn(20,)
#creates an array with 20 random numbers
#mutliplies all of the randomly generated numbers by 100
```

```python
r
#range of the given array
```

```python
plt.plot(r,'.')
#plots the randomized array
```

```python
y2 = m * x + b + r
#plots y 2 as a function of the slope intercept
```

```python
plt.plot(x,y2,'.')
#plotts the x value and the y2 values in the (x,y) order representing each point with a dot
```

```python
m_guess = 1
#sets the guess of the slope of the linear equation to 1
b_guess = 1
#sets the guess of the y intercept of the linear equation to 1

y_guess = m_guess * x + b_guess
#computes the predicted y values

plt.plot(x,y_guess,'r.');
#plots the x, and y guess values with a red dot
plt.plot(x,y2,'g.');
#pots the x and y2 guess values with a green dot
```

```python
for m_guess in np.linspace(-100,100,100):
  #array with 100 points evenly spaced from a range of -100 to 100

        y_guess = m_guess * x
      #calculates predicted values with the current guess for the slope

        plt.plot(x,y_guess,'r.');
        plt.plot(x,y2,'g.');
        plt.show()
        #plotting the predicted values with a red dot for x and y guess
        #plotting the predicted values with a green dor for x and y2

        loss = np.sum((y_guess - y)**2)
        #calculates the sum of the guess and squares that value

        print(loss,'M = ',m_guess)
```

```python
#brute-force search algorithm
best_loss = 100000000000
#sets best loss value to a very large number to ensure that any loss will realistically be smaller than what is computed
best_m = 0
best_b = 0
#intializes the best fitting line with the slope and y intercept to 0

for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):
#array with 100 evenly spaced points ranging from -100 to 100

        y_guess = m_guess * x + b_guess
      #the y value is computed through the slope intercept equation using the x values given as well as the guess m and guess b value

        loss = np.sum((y_guess - y)**2)
        #squares the value of theh y guess minus the y value

        if loss < best_loss:
          #checks if current loss is less than the best loss

            best_loss = loss
            best_m = m_guess
            best_b = b_guess
            #current model is best fitting
```

```python
y_guess = best_m * x + best_b

plt.plot(x,y_guess,'r.');
plt.plot(x,y2,'g.');
plt.show()
#plots the x and y values with red and green dots to show the data in a scatter plot

print(best_loss,best_m,best_b)
```

```python
#approach to linear repression
best_loss_so_far = 1000000000000
#intalizies the variable with a very large number as a placeholder
m_best = 0
#the best loss is the one in which the slope of the graph is 0
b_best = 0
#the best guess is where the slope intercept of the line is equal to 0
for m_guess in np.linspace(-100,100,100):
#100 data plots, evenly spaced from the range of values between -100 and 100 for the slope value
    for b_guess in np.linspace(-100,100,100):
#100 data plots, evenly spaced from the range of values between -100 and 100 for the intercept value

        y_guess = m_guess * X + b_guess
#plots using slope intercept form where y=mx+b

        loss = np.sum((y_guess - Y)**2)
      #calculates the loss between the guesses and the best fit line

        if loss < best_loss_so_far:
            m_best = m_guess
            b_best = b_guess
            best_loss_so_far = loss
#checks if the loss produced is the smallest value loss or if there have been smaller losses found

print("Found best fit:")
print(m_best,b_best,loss
      #print in the x,y axis
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def get_database(url,sheet_name):
#improting libraries
    url_head = "https://docs.google.com/spreadsheets/d/"
    url_foot = "/gviz/tq?tqx=out:csv&sheet="
    url_body = url.split('/')[5]
#defining url
    # sheet_name = "1"
    url_csv = url_head + url_body + url_foot + sheet_name
    df = pd.read_csv(url_csv, sep=',', skiprows=0)
    return df
```
